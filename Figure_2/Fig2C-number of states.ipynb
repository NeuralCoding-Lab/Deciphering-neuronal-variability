```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
probes = ['probeC', 'probeD', 'probeF', 'probeE', 'probeB', 'probeA']
session_ids = [767871931, 771990200, 778240327, 778998620, 786091066,
            794812542, 831882777, 839068429, 847657808, 840012044,
            766640955, 768515987, 771160300, 774875821, 779839471,
            781842082, 787025148, 789848216, 793224716, 816200189,
            819186360, 819701982, 821695405, 829720705, 835479236]
```

```python
# FIGURE 1C
ll = np.load('..\data\LL_eig.npy', allow_pickle=True).item()
sel_metric = np.zeros((len(session_ids), 5))
stim = 'natural_movie_one_more_repeats'
state_range = ll[session_ids[0], stim][:, 0]
for s_no, session_id in enumerate(session_ids):
    sel_metric[s_no] = ll[session_id, stim][:, 1]
    
fig, ax = plt.subplots(figsize=(4.5, 4))
mean = np.mean(sel_metric, axis=0)
sem = np.std(sel_metric, axis=0)
for n in range(len(session_ids)):
    plt.plot(state_range + np.random.rand(1) / 3.5, sel_metric[n, :], marker='o',
             ms=8, ls='none', color='k', mfc='none', lw=0.3)
plt.errorbar(state_range + 0.6, mean, sem, marker='o', color='k', ls='none', ms=10)
plt.xticks(np.arange(2, 7) + 0.3, labels=np.arange(2, 7), fontsize=12);
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_color('k')
ax.spines['left'].set_color('k')
ax.tick_params(bottom=True, left=True, right=False, top=False)
plt.ylabel('Normalized -LL/$\lambda_{1}$', fontsize=14)
plt.ylim(-1, 4)
plt.xticks(fontsize=14)
plt.yticks(np.arange(0, 5), fontsize=14)
plt.xlabel('# states', fontsize=14)
plt.tight_layout()
```
