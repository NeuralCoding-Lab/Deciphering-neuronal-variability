```python
import sys
sys.path.append(r"C:\Users\shailaja.akella\Dropbox (Personal)\variability_phase4\state_dependent_variability_paper\figure_codes")

import seaborn as sns
import numpy as np
from scipy.stats import f_oneway
from einops import rearrange
from scipy.signal import savgol_filter
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

import behavior_utils as bu
import data_utils as du
```

```python
state_names = ['High frequency', 'Intermediate', 'Low frequency']
short_state_names = [r"$S_{H}$", r"$S_{I}$", r"$S_{L}$"]
state_colors = ['tab:green', 'tab:blue', 'mediumvioletred']

probes = ['probeC', 'probeD', 'probeF', 'probeE', 'probeB', 'probeA']
stim = 'natural_movie_one_more_repeats'
Fs = 30
num_states = 3

session_ids = [767871931, 771990200, 778240327, 778998620, 786091066,
            794812542, 831882777, 839068429, 847657808, 840012044,
            766640955, 768515987, 771160300, 774875821, 779839471,
            781842082, 787025148, 789848216, 793224716, 816200189,
            819186360, 819701982, 821695405, 829720705, 835479236]
```

```python
behavior = np.zeros((9, len(session_ids), num_states))

sc = StandardScaler()

for sess_no, session_id in enumerate(session_ids):
    
    # trials
    session = du.get_nwb_session(session_id)
    trials = du.trials_table(session, stim)
    
    # running, pupil, face motion
    duration = int(trials.duration.mean())
    pupil_data = bu.pupil_area(session, stim, trials)
    running_data = bu.running(session, stim, trials)
    binned_running_speed = bu.bin_behavior(running_data, Fs, duration)
    binned_pupil_size = bu.bin_behavior(pupil_data, Fs, duration)
    face_motion = bu.face_motion(session_id = session_id).fit().face_motion
    
    # state
    states = np.load('../data/states/states_' + str(session_id) + '.npy') 
    [n_trials, trial_length] = states.shape
    
    # other behavior markers
    binned_mvmts = np.nan_to_num(bu.pose_tracking_features(session_id, stim).values)
    binned_mvmts = sc.fit_transform(binned_mvmts).T.reshape(-1, n_trials, trial_length)
    face_motion = face_motion.reshape(n_trials, trial_length) 
    
    for s_no in range(num_states):
        behavior[0, sess_no, s_no] = np.nanmean(binned_running_speed[states[:] == s_no])
        behavior[1, sess_no, s_no] = np.nanmean(binned_pupil_size[states[:] ==  s_no])
        behavior[2, sess_no, s_no] = np.nanmean(face_motion[states[:] ==  s_no])
        
        dat = binned_mvmts[:, :states.shape[0]]
        for n in range(binned_mvmts.shape[0]):
            behavior[3+n, sess_no, s_no] = np.nanmean(dat[n, states[:] ==  s_no]) 
```

```python
fig, ax = plt.subplots(3, 3, figsize = (6, 8))
ax = ax.flatten()
labels = ['pupil size', 'velocity', 'face motion',
          'body center',  'left forelimb', 'left hindlimb', 
          'right hindlimb', 'tail start', 'tail end']
for sess_no in range(len(session_ids)):
    for n in range(len(labels)):
        ax[n].plot(np.arange(3), behavior[n, sess_no], color = 'silver', lw = 1)
        for s_no in range(3):
            ax[n].plot(s_no,  behavior[n, sess_no, s_no], mec = 'w', mfc = state_colors[s_no],
                       marker = 'o', ms=8)
        ax[n].spines['right'].set_visible(False)
        ax[n].spines['top'].set_visible(False)
        ax[n].spines['bottom'].set_color('k')
        ax[n].set_ylim(-1.5,1.5)
        if np.mod(n, 3) == 0:
            ax[n].spines['left'].set_color('k')           
            ax[n].set_yticks([-1, 0, 1])
            ax[n].set_ylabel('zscored activity', fontsize = 10)
        else:
            ax[n].spines['left'].set_visible(False)
            ax[n].set_yticks([])
                
        ax[n].set_xlim(-0.3, 2.3)
        ax[n].set_xticks([])
        ax[n].set_xlabel(labels[n])
        ax[n].tick_params(bottom = True,left = True, right = False, top = False)
        
p = np.zeros(len(labels))
for n in range(len(labels)):
    _, p[n] = f_oneway(behavior[n, :, 0][~np.isnan(behavior[n, :, 0])], behavior[n, :, 1:][~np.isnan(behavior[n, :, 1:])] ) 


for i in range(len(labels)):   
    x1, x2 =0, 2   # columns 'Sat' and 'Sun' (first column: 0, see plt.xticks())
    y, h, col = 0.8, 0, 'k'
    if p[i] < 0.0001:   
        ax[i].text(x1, y + 0.2, "***", ha='center', va='bottom', color=col, fontsize = 14)
    elif p[i] < 0.001:
        ax[i].text(x1, y + 0.2, "**", ha='center', va='bottom', color=col, fontsize = 14)
    elif p[i] < 0.05:
        ax[i].text(x1*.5, y + 0.2, "*", ha='center', va='bottom', color=col, fontsize = 14)
    
    
plt.tight_layout()
```

```python
pupil_mi = np.load('../data/mi_behavior_states/pupil_mi.npy')
running_mi = np.load('../data/mi_behavior_states/running_mi.npy')
face_mi = np.load('../data/mi_behavior_states/facemotion_mi.npy')
body_mi = np.load('../data/mi_behavior_states/body_mi.npy')
body_mi = np.delete(body_mi, 3, axis = 0) # remove right forelimb
body_mi = np.delete(body_mi, 0, axis = 0) # remove ears


ind = np.argsort(np.nanmean(body_mi,axis = 1))
y_all = np.concatenate((body_mi[ind],running_mi[0].reshape(1, -1), face_mi[0].reshape(1, -1), pupil_mi[0].reshape(1, -1)))

nodes = ['body\ncenter', 'forelimb(L)', 'hindlimb(L)','hindlimb(R)', 'tail start', 'tail end']
x = np.concatenate((np.array(nodes)[ind], ['running', 'face\nmotion', 'pupil\nsize']))

y = np.concatenate((np.nanmean(body_mi[ind],axis = 1), [np.nanmean(running_mi[0]),np.nanmean(face_mi[0]), np.nanmean(pupil_mi[0])]))
yerr = np.concatenate((np.nanstd(body_mi[ind],axis = 1),[np.nanstd(running_mi[0]),np.nanstd(face_mi[0]), np.nanstd(pupil_mi[0])]))

fig, ax = plt.subplots(figsize = (6.5,4))
plt.errorbar(x, y, yerr/np.sqrt(25), color = 'k', ls = 'none', marker = 'o', ms = 10)
plt.xticks(rotation = 45, fontsize = 12);
plt.yticks([ 0.04, 0.06, 0.08, 0.1, 0.12],fontsize = 12)
plt.ylabel('normalized MI', fontsize = 12)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_color('k')
ax.spines['left'].set_color('k')
plt.grid(True, color = 'silver', alpha = 0.3)

left, bottom, width, height = [0.3, 0.6, 0.4, 0.4]
ax2 = fig.add_axes([left, bottom, width, height])

p = np.ones((len(x), len(x)))*np.nan
for i in range(len(x)):
    for j in range(i+1, len(x)):
        F, p[i, j] = f_oneway(y_all[i][~np.isnan(y_all[i])],y_all[j][~np.isnan(y_all[j])])


nodes = ['body center', 'forelimb(L)', 'hindlimb(L)','hindlimb(R)', 'tail start', 'tail end']
x = np.concatenate((np.array(nodes)[ind], ['running', 'face motion', 'pupil size']))
sns.heatmap(p,yticklabels = x, xticklabels = [], lw = 0.5,ax = ax2)
ax2.set_title('p-values')
plt.xticks(rotation = 45);
plt.tight_layout()
print(p<0.05/55)
```
