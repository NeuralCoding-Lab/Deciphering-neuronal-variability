```python
df_fullmodel = pd.read_csv('../data/HMM_GLM/fullmodel_r2_qc2.csv')
df_behavior =  pd.read_csv('../data/HMM_GLM/behavior_r2_qc2.csv')
df_stim =  pd.read_csv('../data/HMM_GLM/stim_r2_qc2.csv')
df_stim_behavior =  pd.read_csv('../data/HMM_GLM/stim_behavior_r2_qc2.csv')

df_lfp_pop = pd.read_csv('../data/HMM_GLM/lfp_pop_r2_qc2.csv')
df_pop =  pd.read_csv('../data/HMM_GLM/population_r2_qc2.csv')
df_lfp =  pd.read_csv('../data/HMM_GLM/lfp_r2_qc2.csv')
```

```python
df_final = pd.DataFrame()
temp_df = pd.DataFrame()


temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_behavior[df_behavior['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'behavior'
df_final = df_final.append(temp_df, ignore_index = True)

temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_lfp_pop[df_lfp_pop['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'internal\nbrain activity'
df_final = df_final.append(temp_df, ignore_index = True)


temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_stim[df_stim['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'stimulus'
df_final = df_final.append(temp_df, ignore_index = True)

temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_stim_behavior[df_stim_behavior['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'stimulus +\nbehavior'
df_final = df_final.append(temp_df, ignore_index = True)

temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_fullmodel[df_fullmodel['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'full model'
df_final = df_final.append(temp_df, ignore_index = True)


x1 = df_fullmodel[df_fullmodel['final_r2'] > 0]['final_r2'].values
x2 = df_stim_behavior[df_stim_behavior['final_r2'] > 0]['final_r2'].values
F, p = f_oneway(x1,x2)
print(np.nanmean(x1) - np.nanmean(x2))
print(p)

fig, ax = plt.subplots(figsize = (4.2, 3.2))
sns.boxplot(x = 'feature', y = '$r^2$', data = df_final, width = 0.35, 
            palette = 'cubehelix_r', fliersize = 0, medianprops={"color": "w"}, 
            boxprops = {"edgecolor" : 'w'})
plt.ylim(-0.05, 0.7)
plt.yticks([0, 0.2, 0.4, 0.6], fontsize = 12)
plt.xticks(fontsize = 12, rotation = 45)
plt.ylabel('cv$R^2 (> 0)$', fontsize = 12)
plt.xlabel('')
plt.grid(True, color = 'silver', alpha = 0.3)
plt.xticks(fontsize = 11)
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

plt.tight_layout()
```

```python
temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_pop[df_pop['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'other area\npop. activity'
df_final = df_final.append(temp_df, ignore_index = True)


temp_df = pd.DataFrame()
temp_df['$r^2$'] = df_lfp[df_lfp['final_r2'] > 0]['final_r2'].values
temp_df['feature'] = 'same area\nLFPs'
df_final = df_final.append(temp_df, ignore_index = True)
```

```python
stim = 'natural_movie_one_more_repeats'

df_state = pd.DataFrame()
state_df_labels = ['SH_r2', 'SI_r2', 'SL_r2']

for s_no in range(num_states):
    temp_df = pd.DataFrame()
    temp_df['$r^2$'] =  df_fullmodel[df_fullmodel[state_df_labels[s_no]] > 0][state_df_labels[s_no]].values
    temp_df['state'] = short_state_names[s_no]
    temp_df['feature'] = 'full model'
    df_state = df_state.append(temp_df, ignore_index = True)
    
    
for s_no in range(num_states):
    temp_df = pd.DataFrame()
    temp_df['$r^2$'] =  df_stim[df_stim[state_df_labels[s_no]] > 0][state_df_labels[s_no]].values
    temp_df['state'] = short_state_names[s_no]
    temp_df['feature'] = 'stimulus'
    df_state = df_state.append(temp_df, ignore_index = True)

for s_no in range(num_states):
    temp_df = pd.DataFrame()
    temp_df['$r^2$'] = df_pop[df_pop[state_df_labels[s_no]] > 0][state_df_labels[s_no]].values
    temp_df['state'] = short_state_names[s_no]
    temp_df['feature'] = 'other area\npopulation activity'
    df_state = df_state.append(temp_df, ignore_index = True)

    
for s_no in range(num_states):
    temp_df = pd.DataFrame()
    temp_df['$r^2$'] =  df_behavior[df_behavior[state_df_labels[s_no]] > 0][state_df_labels[s_no]].values
    temp_df['state'] = short_state_names[s_no]
    temp_df['feature'] = 'behavior'
    df_state = df_state.append(temp_df, ignore_index = True)
    
for s_no in range(num_states):
    temp_df = pd.DataFrame()
    temp_df['$r^2$'] = df_lfp[df_lfp[state_df_labels[s_no]] > 0][state_df_labels[s_no]].values
    temp_df['state'] = short_state_names[s_no]
    temp_df['feature'] = 'same area\nLFPs'
    df_state = df_state.append(temp_df, ignore_index = True)   
    
fig, ax = plt.subplots(figsize = (10, 3))
sns.boxplot(x = 'feature', y = '$r^2$', hue = 'state', data = df_state, width = 0.5, 
            palette = state_colors, fliersize = 0,  medianprops={"color": "w"}, 
            boxprops = {"edgecolor" : 'w'})
plt.ylim(-0.02, 0.7)
plt.yticks([0, 0.2, 0.4, 0.6], fontsize = 12)
plt.ylabel('cv$R^2$', fontsize = 12)
plt.xticks(fontsize = 12)
plt.xlabel('')
plt.grid(True, zorder = 0,alpha= 0.3, color = 'silver')
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

plt.legend(bbox_to_anchor = (1.15, 1), fontsize = 12) 
plt.tight_layout()
```

```python
cluster_names = ['unexplained','stimulus', 'behavior', 'stimulus+behavior','stimulus+population',   'multi-source']
colors = sns.color_palette('rocket_r', len(cluster_names) + 8)[:-2][::2]
final_colors = dict(zip(cluster_names, colors))

def clean_cluster_df(df): 
    df['cluster_name'] = df['cluster_name'].str.replace(r'\d+', '', regex=True)
    df['probe'] = np.round(df['probe'], 3)
    return df

df_final = pd.read_csv('../data/HMM_GLM/clustering/clustering_with_names.csv')
df_SH = pd.read_csv('../data/HMM_GLM/clustering/clustering_with_names_SH.csv')
df_SL = pd.read_csv('../data/HMM_GLM/clustering/clustering_with_names_SL.csv')

df_final = clean_cluster_df(df_final) 
df_SH = clean_cluster_df(df_SH) 
df_SL = clean_cluster_df(df_SL) 

df_SH['cluster_name'] = df_SH['cluster_name'].str.replace('multisource', 'multi-source')
df_SL['cluster_name'] = df_SL['cluster_name'].str.replace('multisource', 'multi-source')
```

```python
fig, ax = plt.subplots(figsize = (5, 5))
sizes = [len(df_final[df_final.cluster_name == cluster_name])/len(df_final) for cluster_name in cluster_names]
ax.pie(sizes, labels = cluster_names, autopct='%1.1f%%',)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 3), sharex = True, sharey = True)
prev = np.zeros(6)
NC = len(final_colors)

for cluster_name in final_colors.keys():
    counts,bins = np.histogram(df_final[df_final.cluster_name == cluster_name].probe, bins = 6)
    L = np.array([counts[i]/len(df_final[df_final.probe == np.round(i/6,3)]) for i in range(6)])
    name = cluster_name + ' (' + str(len(df_final[df_final.cluster_name == cluster_name])) + ')' 
    ax[0].bar(names, L, bottom = prev, color = final_colors[cluster_name], label = name)
    prev = L + prev
    ax[0].set_ylabel('proportion of units')
    
prev = np.zeros(6)
for cluster_name in final_colors.keys():
    counts,bins = np.histogram(df_SH[df_SH.cluster_name == cluster_name].probe, bins = 6)
    L = np.array([counts[i]/len(df_SH[df_SH.probe == np.round(i/6,3)]) for i in range(6)])
    name = cluster_name + ' (' + str(len(df_SH[df_SH.cluster_name == cluster_name])) + ')' 
    ax[1].bar(names, L, bottom = prev, color = final_colors[cluster_name], label = name)
    prev = L + prev
    
    
prev = np.zeros(6)
for cluster_name in final_colors.keys():
    counts,bins = np.histogram(df_SL[df_SL.cluster_name == cluster_name].probe, bins = 6)
    L = np.array([counts[i]/len(df_SL[df_SL.probe == np.round(i/6,3)]) for i in range(6)])
    name = cluster_name + ' (' + str(len(df_SL[df_SL.cluster_name == cluster_name])) + ')' 
    ax[2].bar(names, L, bottom = prev, color = final_colors[cluster_name], label = name)
    prev = L + prev
    
for i in range(3): 
    ax[i].spines["top"].set_visible(False)
    ax[i].spines["right"].set_visible(False)
ax[0].legend(bbox_to_anchor = (1.05, 1))
ax[1].legend(bbox_to_anchor = (1.05, 1))
ax[2].legend(bbox_to_anchor = (1.05, 1))

plt.tight_layout()
# plt.savefig(out_folder + '/r2_cluster_area_distributions_all.pdf')
```

```python
HS_no_RL = [-0.5, -0.13, 0, 0.12, 0.29]
HS = [-0.5, -0.13, -0.14, 0, 0.12, 0.29]


print('------------------------------------------')
print('Correlation with hierarchy score including RL') 
print('------------------------------------------')

for cluster_name in final_colors.keys(): 
    counts = [len(df_final[(df_final.cluster_name == cluster_name) & (df_final.probe == np.round(i/6,3))])/len(df_final[df_final.probe == np.round(i/6,3)]) for i in [0, 1, 3, 4, 5]]
    print(cluster_name + '\nProportions: ', end = '')
    print(np.round(counts,2))
    r, p =  pearsonr(counts, HS_no_RL)
    print('Correaltion: ' + str(np.round(r,2)) + ', p = ' + str(p))

print('------------------------------------------')
print('Correlation with hierarchy score without RL') 
print('------------------------------------------')

for cluster_name in final_colors.keys(): 
    counts = [len(df_final[(df_final.cluster_name == cluster_name) & (df_final.probe == np.round(i/6,3))])/len(df_final[df_final.probe == np.round(i/6,3)]) for i in [0, 1, 2, 3, 4, 5]]
    print(cluster_name + '\nProportions:', end = '')
    print(np.round(counts,2))
    r, p =  pearsonr(counts, HS)
    print('Correaltion: ' + str(np.round(r,2)) + ', p = ' + str(p))
```
